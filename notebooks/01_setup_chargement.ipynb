```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid", palette="muted")
plt.rcParams["figure.figsize"] = (12, 6)

df = pd.read_csv("../data/creditcard.csv")

print(f"Lignes : {df.shape[0]:,}  |  Colonnes : {df.shape[1]}")
print("\n", df.head())
```

```python
# Analyse du déséquilibre
counts = df["Class"].value_counts()
pct_fraud = counts[1] / len(df) * 100

print(f"Transactions légitimes  : {counts[0]:,}  ({100 - pct_fraud:.2f}%)")
print(f"Transactions frauduleuses: {counts[1]:,}  ({pct_fraud:.4f}%)")

# Visualisation
fig, ax = plt.subplots()
ax.bar(["Legitime", "Fraude"], counts.values,
       color=["#378ADD", "#E24B4A"], edgecolor="none", width=0.4)
ax.set_title("Desequilibre des classes - donnees brutes", fontsize=14)
ax.set_ylabel("Nombre de transactions")
for i, v in enumerate(counts.values):
    ax.text(i, v + 1000, f"{v:,}", ha="center", fontsize=11)
plt.tight_layout()
plt.savefig("../outputs/01_desequilibre_classes.png", dpi=150)
plt.show()
```
